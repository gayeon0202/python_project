```python
df_score = df.loc[:, '국어':'사회']
df_score['총점'] = df_score.apply(lambda row:sum(row['국어':'사회']), axis=1)
df_score['평균']= df_score.apply(lambda row:row['총점']/5, axis=1)
df_score.loc['평균'] = df_score.apply(np.mean, axis=0)
df_score
```
